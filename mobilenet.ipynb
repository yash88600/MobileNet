```python
import numpy
import pandas
import keras
from keras.layers import Conv2D,AveragePooling2D,SeparableConv2D,Dense,Flatten,Activation,Input,BatchNormalization,DepthwiseConv2D,Dropout
from keras.models import Model
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_path='/kaggle/input/animals10/raw-img'
```

```python
train_datagen = ImageDataGenerator(rescale=1./255,validation_split=0.2)
train_batch = train_datagen.flow_from_directory(train_path,target_size=(128, 128),batch_size=128,subset='training')
validation_batch = train_datagen.flow_from_directory(train_path,target_size=(128, 128),batch_size=128,subset='validation')
```

```python
type(train_batch)
```

```python
X_input=Input((128,128,3))
```

```python

#First layer(standard conv layer)
x=Conv2D(32,(3,3),strides=2,padding='SAME')(X_input)
x=BatchNormalization()(x)
x=Activation('relu')(x)

#Second layer(Depthwise separable conv layer)
x=DepthwiseConv2D((3,3),strides=1,padding='SAME')(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)
x=Conv2D(64,(1,1),strides=1)(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)

#Third layer(Depthwise separable conv layer)
x=DepthwiseConv2D((3,3),strides=2,padding='SAME')(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)
x=Conv2D(128,(1,1),strides=1)(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)

#Fourth layer(Depthwise separable conv layer)
x=DepthwiseConv2D((3,3),strides=1,padding='SAME')(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)
x=Conv2D(128,(1,1),strides=1)(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)

#Fifth layer(Depthwise separable conv layer)
x=DepthwiseConv2D((3,3),strides=2,padding='SAME')(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)
x=Conv2D(256,(1,1),strides=1)(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)

#Sixth layer(Depthwise separable conv layer)
x=DepthwiseConv2D((3,3),strides=1,padding='SAME')(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)
x=Conv2D(256,(1,1),strides=1)(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)

#Seventh layer(Depthwise separable conv layer)
x=DepthwiseConv2D((3,3),strides=2,padding='SAME')(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)
x=Conv2D(512,(1,1),strides=1)(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)

#Eighth layer(Depthwise separable conv layer)
x=DepthwiseConv2D((3,3),strides=1,padding='SAME')(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)
x=Conv2D(512,(1,1),strides=1)(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)

#Ninth layer(Depthwise separable conv layer)
x=DepthwiseConv2D((3,3),strides=1,padding='SAME')(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)
x=Conv2D(512,(1,1),strides=1)(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)

#Tenth layer(Depthwise separable conv layer)
x=DepthwiseConv2D((3,3),strides=1,padding='SAME')(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)
x=Conv2D(512,(1,1),strides=1)(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)

#Eleventh layer(Depthwise separable conv layer)
x=DepthwiseConv2D((3,3),strides=1,padding='SAME')(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)
x=Conv2D(512,(1,1),strides=1)(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)

#Twelfth layer(Depthwise separable conv layer)
x=DepthwiseConv2D((3,3),strides=1,padding='SAME')(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)
x=Conv2D(512,(1,1),strides=1)(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)



#Thirtheenth layer(Depthwise separable conv layer)
x=DepthwiseConv2D((3,3),strides=2,padding='SAME')(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)
x=Conv2D(1024,(1,1),strides=1)(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)

#Fourteenth layer(Depthwise separable conv layer)
x=DepthwiseConv2D((3,3),strides=1,padding='SAME')(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)
x=Conv2D(1024,(1,1),strides=1)(x)
x=BatchNormalization()(x)
x=Activation('relu')(x)

#Average Pooling
x=AveragePooling2D((4,4))(x)

#Fully Connected Layer
x=Flatten()(x)
x=Dense(10)(x)
x=Activation('softmax')(x)
```

```python
x.shape
```

```python
mobilenet=Model(inputs=X_input,outputs=x)
```

```python
mobilenet.summary()
```

```python
opt=keras.optimizers.RMSprop(learning_rate=0.01/5)
mobilenet.compile(optimizer=opt,loss='categorical_crossentropy',metrics=["accuracy"])
```

```python
mobilenet.fit_generator(train_batch,epochs=50,validation_data=validation_batch)
```

```python
mobilenet.save('mobile.h5')
```
